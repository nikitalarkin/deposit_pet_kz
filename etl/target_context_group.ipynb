```python
import pandas as pd
import sqlalchemy as sa
```

```python
context_kz = pd.read_csv('../clean_data/contextKZ.csv', parse_dates=['month'])
target_full = pd.read_csv('../clean_data/target.csv', parse_dates=['month'])
target = target_full[target_full.iloc[:, 0] == 'Всего по республике:'].reset_index(drop=True)
```

```python
macro_kz = context_kz.merge(target, on='month', how='inner').drop(columns=['region'])
```

```python
engine = sa.create_engine("postgresql+psycopg2://postgres:qwerty123@localhost:5432/credits_total_kz")
```

```python
context_kz.to_sql('context_kz', engine, if_exists='replace', index=False, dtype={'month': sa.DateTime()})
target_full.to_sql('target', engine, if_exists='replace', index=False, dtype={'month': sa.DateTime()})
macro_kz.to_sql('macro_kz', engine, if_exists='replace', index=False, dtype={'month': sa.DateTime()})
```
